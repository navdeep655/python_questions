### Q8. Filter palindrome strings using filter().

words = ["madam", "python", "level", "hello", "radar"]

```python
words = ["madam", "python", "level", "hello", "radar"]
x=filter(lambda x:x ==x[::-1],words)
print(list(x))
```

#### Q9. Write a lambda function that returns the maximum of two numbers.

Example:

max(10, 20) → 20

```python
x=lambda a,b:b if b>a else a
print( x(10,20))
```

```python
def fun():
    return lambda x,y:y>x
z=fun()
print(z(10,20))
```

### Q13. Given: 
nums = [5, 12, 17, 18, 24, 32]

Return the sum of squares of even numbers using filter() and map().

```python
nums = [5, 12, 17, 18, 24, 32]
k=filter(lambda x:x%2==0,nums)
def fun():
    return list(map (lambda x:x**2,k))
sum(fun())
```

### Q15. Given dictionary:

marks = {"Aman":80,"Riya":45,"Karan":60,"Neha":30}

Create a new dictionary containing only students with marks ≥ 50.

```python
marks = {"Aman":80,"Riya":45,"Karan":60,"Neha":30}
new={}
for key,value in marks.items():
    if value>=50:
        new.update({key:value})
list(new.keys())
```
