## 1. Create a Statement that will print out words that start with 's':
str_1 = 'Print only the words that start with s in this sentence Said Ronney'

```python
str1='Print only the words that start with s in this sentence Said Ronney'
str2=str1.lower()
str3=str2.split()
print(str3)
#print(str3[5][0]=="s")
for i in range(len(str3)):
        if str3[i][0]=="s":
            print(str3[i])
```

## 4. Go through the string below and if the length of a word is even print "even!"
st = 'Print every word in this sentence that has an even number of letters'

```python
st = 'Print every word in this sentence that has an even number of letters'
st1=st.split()
for i in range(len(st1)):
    if len(st1[i])%2==0:
        print("even")
    else:
        print(st1[i])
        
```

## 7. Write a program to replace the second occurrence of a character in a string with "?" if the next character is the same as the current character.
st = 'AFDFFAAAOPPEMHSSOIS'

```python
st = 'AFDFFAAAOPPEMHSSOIS'
result=""
for i in range(len(st)):
    if st[i-1]==st[i]:
        result+="?"
    else:
        result+=st[i]
result
```

## 8. Write a program to count the number of vowels in a string.
input_string =  "programming"

```python
input_string = "programming"
vowels="aieouAEIOU"
count=0
for i in range(len(vowels)):
    for j in range(len(input_string)):
        if vowels[i]==input_string[j]:
            count+=1
count
```
